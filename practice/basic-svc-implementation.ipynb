# Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Lets create a synthetic data points
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
## Prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## RBF Kernel

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
svc_rbf = SVC(kernel='rbf')
```

```python
svc_rbf.fit(X_train, y_train)
```

```python
## Prediction
y_pred = svc_rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## Polynomial

```python
svc_polynomial = SVC(kernel='poly')
svc_polynomial.fit(X_train, y_train)
## Prediction
y_pred = svc_polynomial.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## Sigmoid

```python
svc_sigmoid = SVC(kernel='sigmoid')
svc_sigmoid.fit(X_train, y_train)
## Prediction
y_pred = svc_sigmoid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## Hyperparemeter tuning

```python
from sklearn.model_selection import GridSearchCV

# Defining parameter range
param_grid = {
    'C': [0.1, 1, 10, 100, 1000], 
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
    'kernel': ['rbf']
}
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=5, verbose=3, refit=True)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
## Prediction
y_pred = grid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
