# Stopwords

These are words which stops, joins or continue sentences, like to, and, she, of, etc. These kind of words are not required for the machine or deep learning algorithms since they don't provide any value for classification or prediction. The reason we do this is because we need to convert the important words into numbers which can be used on mathematical equations.

```python
corpus = """Monkey D. Luffy, also known as "Straw Hat" Luffy, 
            is a fictional character and the protagonist in the Japanese manga series One Piece created by Eiichiro Oda.
            Luffy made his debut as a young boy who acquires the properties of rubber after accidentally eating one of the Devil Fruits
            that belonged to "Red Hair" Shanks. Monkey D. Luffy is the captain of the Straw Hat Pirates, and dreamt of being a pirate since
            childhood from the influence of his idol and mentor Red-Haired Shanks. At the age of 17, Luffy sets sail from the East Blue Sea
            to the Grand Line in search of the legendary treasure, One Piece, to succeed Gol D. Roger as "King of the Pirates".
            He fights multiple antagonists, and aids and befriends the inhabitants of several islands on his journey. Usually cheerful,
            he becomes serious and even aggressive when he fights. Luffy uses his rubber body to concentrate his power, executing a range
            of attacks.
"""
```

```python
print(corpus)
```

```python
from nltk.stem import SnowballStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words('spanish')
```

```python
stemmer = SnowballStemmer('english')
```

```python
sentences = nltk.sent_tokenize(corpus)
print(sentences)
```

```python
## Apply Stopwords and Filter and then Apply Stemming

local_stopwords = set(stopwords.words('english'))

def get_words_from(sentence):
    return nltk.word_tokenize(sentence)

def apply_stemming_if_not_stopword(words):
    return [stemmer.stem(word) for word in words if word not in local_stopwords]

def convert_words_into_setence(words):
    return ' '.join(words)

for i in range(len(sentences)):
    words = get_words_from(sentences[i])
    words = apply_stemming_if_not_stopword(words)
    sentences[i] = convert_words_into_setence(words)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
## Apply Stopwords and Filter and then Apply Stemming

local_stopwords = set(stopwords.words('english'))

def get_words_from(sentence):
    return nltk.word_tokenize(sentence)

def apply_stemming_if_not_stopword(words):
    return [lemmatizer.lemmatize(word.lower(), pos='v') for word in words if word not in local_stopwords]

def convert_words_into_setence(words):
    return ' '.join(words)

for i in range(len(sentences)):
    words = get_words_from(sentences[i])
    words = apply_stemming_if_not_stopword(words)
    sentences[i] = convert_words_into_setence(words)
```
